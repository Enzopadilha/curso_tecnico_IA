## 1. Revisão rápida: o que é logaritmo?

O logaritmo responde à pergunta:

> **“A que potência precisamos elevar uma base para obter determinado valor?”**

Exemplo:


$$\log_2(8) = 3$$


Isso significa que:

$$
2^3 = 8
$$

---

## Ideia central

Logaritmo e exponencial são operações inversas.

Se:

$$
2^3 = 8
$$

então:

$$
\log_2(8) = 3
$$

---

## Forma geral

$$
f(x) = \log_b(x)
$$

onde:

- `b` é a base do logaritmo;
- `x` é o valor de entrada;
- `f(x)` é o resultado.

## 2. Intuição prática

Funções logarítmicas aparecem quando os valores crescem muito rápido e precisamos “comprimir” a escala para enxergar melhor os dados.

### Exemplos do dia a dia:
- escala Richter de terremotos;
- intensidade sonora em decibéis;
- crescimento de dados em tecnologia;
- comparação entre valores muito diferentes.

### Ideia importante:
O logaritmo cresce, mas cresce **devagar**.

Isso é o oposto da função exponencial, que pode crescer muito rapidamente.

## 3. Exemplo básico

Vamos calcular alguns valores da função:

$$
f(x) = \log_2(x)
$$

Lembrando que só podemos calcular logaritmo de valores **positivos**.

```python
import numpy as np

valores = [1, 2, 4, 8, 16, 32]
[(x, np.log2(x)) for x in valores]
```

## Resolução comentada do exemplo

- $$log_2(1) = 0$$, porque \(2^0 = 1\)
\
- $$log_2(2) = 1$$, porque \(2^1 = 2\)
  
- $$log_2(4) = 2$$, porque \(2^2 = 4\)
  
- $$log_2(8) = 3$$, porque \(2^3 = 8\)

- $$log_2(16) = 4$$, porque \(2^4 = 16\)  

\
Perceba que, mesmo dobrando muito o valor de entrada, a saída cresce lentamente.

## 4. Relação entre exponencial e logaritmo

Vamos comparar:

### Exponencial
$$
2^x
$$

### Logarítmica
$$
\log_2(x)
$$

Uma cresce muito rápido.  
A outra cresce devagar.

Isso é essencial para entender dados, algoritmos e IA.

```python
import matplotlib.pyplot as plt

x_exp = np.linspace(0, 6, 200)
y_exp = 2 ** x_exp

x_log = np.linspace(0.1, 20, 400)
y_log = np.log2(x_log)

plt.figure(figsize=(8, 5))
plt.plot(x_exp, y_exp, label="2^x")
plt.plot(x_log, y_log, label="log2(x)")
plt.title("Comparação entre função exponencial e logarítmica")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid()
plt.show()
```

```python
x_log = np.linspace(0.1, 20, 400)
y_log = np.log2(x_log)

plt.figure(figsize=(8, 5))
plt.plot(x_log, y_log, label="log2(x)")
plt.title("Função logarítmica")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Valores para a função exponencial
x_exp = np.linspace(-3, 3, 400)
y_exp = 2 ** x_exp

# Valores para a função logarítmica
x_log = np.linspace(0.1, 8, 400)
y_log = np.log2(x_log)

# Valores para a reta y = x
x_reta = np.linspace(-3, 8, 400)
y_reta = x_reta

# Criando o gráfico
plt.figure(figsize=(8, 6))

plt.plot(x_exp, y_exp, label='y = 2^x')
plt.plot(x_log, y_log, label='y = log2(x)')
plt.plot(x_reta, y_reta, '--', label='y = x')

# Eixos
plt.axhline(0)
plt.axvline(0)

plt.xlim(-3, 8)
plt.ylim(-4, 8)

plt.title('Simetria entre a função exponencial e logarítmica')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)

plt.show()
```

## 7. Mostrando um exemplo ligado à IA e análise de dados

Exemplo:
- 10

Em vez de olhar esses valores na escala normal, podemos aplicar logaritmo para enxergar melhor as diferenças.

Crie uma aplicação simples de logaritmo para uma das situações abaixo:

3. transformação de dados grandes em escala logarítmica;

Depois:

```python
# Problema para redução de etapas
# temos 100 problemas e a cada etapa reduzimos pela metade

problema = 100
etapas = [0]
valores = [problema]
etapa = 0

while problema > 1:
    problema = problema / 2
    etapa += 1
    etapas.append(etapa)
    valores.append(problema)
    print(f"Etapa {etapa}: Tamanho do problema = {problema:.2f}")
```

```python
import matplotlib.pyplot as plt

# Criando o gráfico
plt.figure(figsize=(8, 5))
plt.plot(etapas, valores, marker='o', color='red', label='Tamanho do Problema')

# Customização
plt.title("Redução Logarítmica (Divisão por 2)")
plt.xlabel("Etapa")
plt.ylabel("Tamanho do Problema")
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend()

plt.show()
```

## Resposta do desafio 1
O(log(n)) é o tempo de complexidade que descreve algoritmos que reduzem o tempo pela metade. Quando *n* aumenta, o número de operações cresce de forma logaritma. Concluindo, significa dizer que o tempo de execução desse algoritmo cresce lentamente.

```python
# Resposta do desafio 2

import numpy as np
import matplotlib.pyplot as plt

# Valores para a função exponencial
x_exp = np.linspace(-3, 5, 400)
f_x = 2 ** x_exp

# Valores para a função logarítmica
x_log = np.linspace(0.1, 32, 400)
g_log = np.log2(x_log)

plt.figure(figsize=(10, 6))

plt.plot(x_exp, f_x, label='f(x) = 2^x', color='blue')
plt.plot(x_log, g_log, label='g(x) = log2(x)', color='red')

# Eixo y=x para mostrar a simetria (funções inversas)
x_reta = np.linspace(-3, 32, 400)
y_reta = x_reta
plt.plot(x_reta, y_reta, '--', color='gray', label='y = x')

plt.title('Comparação entre Função Exponencial e Logarítmica')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)

# Limites para melhor visualização
plt.xlim(-4, 10)
plt.ylim(-4, 10)

plt.show()

# Respostas às perguntas:
# - qual cresce mais rápido; Resposta: A função exponencial (2^x) cresce muito mais rápido.

# - qual cresce mais devagar; Resposta: A função logarítmica (log2(x)) cresce muito mais devagar.

# - por que isso é importante em IA e computação. Resposta: Em IA e computação, o crescimento logarítmico
# é desejável para algoritmos, pois indica que o tempo de execução ou o uso de recursos aumenta muito
# lentamente com o tamanho da entrada, tornando-os eficientes. Já o crescimento exponencial é geralmente
# evitado para o tempo de execução de algoritmos, pois rapidamente se torna impraticável para grandes entradas.
```

```python
# Desafio 3

# Contando o número de operações
import math

def busca_linear(n):
    operacoes = 0
    for i in range(n):
        operacoes += 1
    return operacoes

def busca_binaria(n):
    operacoes = 0
    tamanho = n
    while tamanho > 1:
        tamanho = tamanho // 2
        operacoes += 1
    return operacoes

# Testando com diferentes tamanhos
valores = [10, 100, 1000, 10000, 100000]

print(f"{'N':>10} | {'Linear':>10} | {'Log (Binária)':>15}")
print("-" * 40)

for n in valores:
    lin = busca_linear(n)
    log = busca_binaria(n)
    print(f"{n:>10} | {lin:>10} | {log:>15}")


# A função logaritmica é melhor quando se quer eficiencia em grandes volumes de dados
```
