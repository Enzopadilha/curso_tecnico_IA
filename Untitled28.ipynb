```python

import pandas as pd
import xml.etree.ElementTree as ET
from google.colab import files
```

```python
# Upload dos XMLs
uploaded = files.upload()

ns = {'nfe': 'http://www.portalfiscal.inf.br/nfe'}

```

```python
#  Esta função é usada para extrair o texto de um elemento específico dentro de um XML
def texto(elemento, caminho, default=''):
    achado = elemento.find(caminho, ns)
    return achado.text if achado is not None and achado.text is not None else default

def numero(valor):
    try:
        return float(valor)
    except:
        return 0

```

```python
linhas = []

for nome_arquivo in uploaded.keys():
    tree = ET.parse(nome_arquivo)
    root = tree.getroot()

    infNFe = root.find('.//nfe:infNFe', ns)

    # Dados gerais da nota
    chave_nfe = infNFe.attrib.get('Id', '').replace('NFe', '')
    numero_nf = texto(infNFe, 'nfe:ide/nfe:nNF')
    serie = texto(infNFe, 'nfe:ide/nfe:serie')
    data_emissao = texto(infNFe, 'nfe:ide/nfe:dhEmi')
    natureza_operacao = texto(infNFe, 'nfe:ide/nfe:natOp')

    emitente_cnpj = texto(infNFe, 'nfe:emit/nfe:CNPJ')
    emitente_nome = texto(infNFe, 'nfe:emit/nfe:xNome')

    destinatario_cnpj = texto(infNFe, 'nfe:dest/nfe:CNPJ')
    destinatario_nome = texto(infNFe, 'nfe:dest/nfe:xNome')

    valor_total_nf = numero(texto(infNFe, 'nfe:total/nfe:ICMSTot/nfe:vNF'))
    valor_total_produtos = numero(texto(infNFe, 'nfe:total/nfe:ICMSTot/nfe:vProd'))

    # Cada item/produto da nota
    for det in infNFe.findall('nfe:det', ns):
        prod = det.find('nfe:prod', ns)
        imposto = det.find('nfe:imposto', ns)

        item = {
            'arquivo_origem': nome_arquivo,
            'chave_nfe': chave_nfe,
            'numero_nf': numero_nf,
            'serie': serie,
            'data_emissao': data_emissao,
            'natureza_operacao': natureza_operacao,

            'emitente_cnpj': emitente_cnpj,
            'emitente_nome': emitente_nome,
            'destinatario_cnpj': destinatario_cnpj,
            'destinatario_nome': destinatario_nome,

            'valor_total_nf': valor_total_nf,
            'valor_total_produtos': valor_total_produtos,

            'n_item': det.attrib.get('nItem'),
            'codigo_produto': texto(prod, 'nfe:cProd'),
            'ean': texto(prod, 'nfe:cEAN'),
            'produto': texto(prod, 'nfe:xProd'),
            'ncm': texto(prod, 'nfe:NCM'),
            'cest': texto(prod, 'nfe:CEST'),
            'cfop': texto(prod, 'nfe:CFOP'),
            'unidade': texto(prod, 'nfe:uCom'),
            'quantidade': numero(texto(prod, 'nfe:qCom')),
            'valor_unitario': numero(texto(prod, 'nfe:vUnCom')),
            'valor_produto': numero(texto(prod, 'nfe:vProd')),

            'lote': texto(prod, 'nfe:rastro/nfe:nLote'),
            'qtd_lote': numero(texto(prod, 'nfe:rastro/nfe:qLote')),
            'data_fabricacao': texto(prod, 'nfe:rastro/nfe:dFab'),
            'data_validade': texto(prod, 'nfe:rastro/nfe:dVal'),

            'codigo_anvisa': texto(prod, 'nfe:med/nfe:cProdANVISA'),
            'valor_pmc': numero(texto(prod, 'nfe:med/nfe:vPMC')),

            'icms_base': numero(texto(imposto, 'nfe:ICMS/*/nfe:vBC')),
            'icms_percentual': numero(texto(imposto, 'nfe:ICMS/*/nfe:pICMS')),
            'icms_valor': numero(texto(imposto, 'nfe:ICMS/*/nfe:vICMS')),

            'pis_valor': numero(texto(imposto, 'nfe:PIS/*/nfe:vPIS')),
            'cofins_valor': numero(texto(imposto, 'nfe:COFINS/*/nfe:vCOFINS')),

            'info_adicional_produto': texto(det, 'nfe:infAdProd')
        }

        linhas.append(item)

df = pd.DataFrame(linhas)

```
