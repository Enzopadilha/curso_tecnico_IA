## 1. Função Linear

### Forma geral

$$
f(x) = ax + b
$$

### Ideia principal
A função linear representa uma relação de crescimento ou decrescimento constante.

- `a` = taxa de variação
- `b` = valor inicial

### Exemplo
$$
f(x) = 2x + 3
$$

Se `x = 4`:

$$
f(4) = 2 \cdot 4 + 3 = 11
$$

### Onde isso aparece em IA?
- regressão linear;
- previsão de valores;
- interpretação de relação entre entrada e saída.

```python
# Exemplo de função linear - a mais básica

def funcao_linear(x):
  return 5*x + 33

lista = [0, 2, 4, 8]
for x in lista:
  print(f'f({x}) -> {funcao_linear(x)}')
```

### Exemplo prático
Vamos supor que uma corrida no aplicativo Uber é cobrada por um valor fixo de R\$5,00 e mais R\$1,00 por KM rodado. Construa uma função que represente a situação e calcula o valor de uma corrida de 15km.

```python
def exemplo(x):
  return x + 5

print(f'f({15}) -> {exemplo(15)}')
```

## 2. Função Polinomial

### Exemplo mais importante na UC3
Função do 2º grau:

$$
f(x) = ax^2 + bx + c
$$

### Ideia principal
A função polinomial permite representar comportamentos curvos.

### Exemplo
$$
f(x) = x^2 - 4x + 5
$$

### Observações
- se `a > 0`, a parábola abre para cima;
- se `a < 0`, a parábola abre para baixo.

### Onde isso aparece em IA?
- modelagem de relações não lineares;
- transformação de variáveis (`x²`, `x³`);
- regressão polinomial.

```python
# Exemplo de função polinomial
def poli(x):
  return x**2 - (4*x) + 5

valores = [1, 5, 8, 9]
for x in valores:
  print(f'f({x}) -> {poli(x)}')
```

## 3. Função Exponencial

### Forma geral

$$
f(x) = a \cdot b^x
$$

### Ideia principal
Na função exponencial, o crescimento ou decrescimento acontece por multiplicação constante.

### Exemplo
$$
f(x) = 3 \cdot 2^x
$$

### Observações
- se `b > 1`, há crescimento;
- se `0 < b < 1`, há decrescimento.

### Onde isso aparece em IA?
- função sigmoid;
- probabilidades;
- crescimento de combinações;
- comportamento de certos algoritmos.

```python
# Exemplo de função exponencial
def funcao_expo(x):
  return 3*(2**x)

for x in valores:
  print(f'f({x}) -> {funcao_expo(x)}')
```

## 4. Função Logarítmica

### Forma geral

$$
f(x) = \log_b(x)
$$

### Ideia principal
O logaritmo responde: “a que potência a base deve ser elevada para obter esse valor?”

### Exemplo
$$
\log_2(8) = 3
$$

porque:

$$
2^3 = 8
$$

### Onde isso aparece em IA?
- entropia;
- funções de perda;
- análise de escalas;
- complexidade de algoritmos.

```python
# Exemplo de logaritmo

import numpy as np

valores = [1,2,3,4]
for x in valores:
  print(np.log2(x))
```

## 5. Probabilidade Básica

### Fórmula

$$
P(A) = \frac{\text{casos favoráveis}}{\text{casos possíveis}}
$$

### Ideias principais
- probabilidade mede chance;
- eventos independentes: multiplicamos probabilidades;
- “e” é diferente de “ou”.

### Exemplos
- probabilidade de cara em uma moeda: \(1/2\)
- probabilidade de cara duas vezes seguidas: \((1/2) \cdot (1/2) = 1/4\)

### Onde isso aparece em IA?
- classificação;
- decisão baseada em probabilidades;
- Naive Bayes;
- previsão com incerteza.

```python
# Exemplo simples de decisão da IA com probabilidades

classes = ["cachorro", "gato", "onça"]
probabilidades = [0.2, 0.6, 0.2]

maior_prob = probabilidades[0]
indice_maior = 0

for i in range(len(probabilidades)):
    if probabilidades[i] > maior_prob:
        maior_prob = probabilidades[i]
        indice_maior = i

decisao = classes[indice_maior]

print("Probabilidades:", probabilidades)
print("Decisão da IA:", decisao)
```

## Bloco E — Probabilidade

### Exercício 17
Qual a probabilidade de sair número par em um dado?

```python
# Total de resultados possíveis
total_resultados = 6

# Os números pares em um dado são: 2, 4, 6
resultados_favoraveis = 3

# Calcula a probabilidade
probabilidade = resultados_favoraveis / total_resultados

print(f"A probabilidade de sair um número par em um dado é: {probabilidade:.2f} ({probabilidade*100:.0f}%)")
```

### Exercício 18
Qual a probabilidade de sair cara duas vezes seguidas ao lançar uma moeda duas vezes?

```python
# Probabilidade de sair cara no primeiro lançamento
prob_cara_1 = 0.5

# Probabilidade de sair cara no segundo lançamento
prob_cara_2 = 0.5

# A probabilidade de sair cara duas vezes seguidas é o produto das probabilidades individuais
prob_cara_duas_vezes = prob_cara_1 * prob_cara_2

print(f"A probabilidade de sair cara duas vezes seguidas é: {prob_cara_duas_vezes:.2f} ({prob_cara_duas_vezes*100:.0f}%)")
```

### Exercício 19
Ao lançar uma moeda e um dado, qual a probabilidade de sair:

1. cara e número 6  
2. cara ou número 6

```python
# Probabilidade de sair cara na moeda
probabilidade_cara = 0.5

# Probabilidade de sair número 6 no dado
probabilidade_6 = 1/6

# Proababilidade de ser cara E número 6 (multiplicação)
prob_cara_e_6 = probabilidade_cara * probabilidade_6

print(f"A probabilidade de sair cara e número 6 é: {prob_cara_e_6:.2f} ({prob_cara_e_6*100:.0f}%)")
```

```python
# Probabilidade de sair cara na moeda
probabilidade_cara = 0.5

# Probabilidade de sair número 6 no dado
probabilidade_6 = 1/6

# Proababilidade de ser cara OU número 6 (soma)
prob_cara_ou_6 = probabilidade_cara + probabilidade_6

print(f"A probabilidade de sair cara ou número 6 é: {prob_cara_ou_6:.2f} ({prob_cara_ou_6*100:.0f}%)")
```


### Desafio 2
Crie um pequeno sistema que receba probabilidades de classes e escolha a decisão final da IA.

```python
classes_aves = ['pato', 'ganso', 'marreco']
probabilidades = [0.2, 0.8, 0.5]

# maior valor na lista
valor_maximo = max(probabilidades)

# índice da probabilidade máxima
indice_decisao = probabilidades.index(valor_maximo)

# Identifica a classe correspondente usando o índice
decisao_final_ia = classes_aves[indice_decisao]

print(f"Lista de probabilidades: {probabilidades}")
print(f"Maior probabilidade: {valor_maximo}")
print(f"Índice da decisão (maior probabilidade): {indice_decisao}")
print(f"A decisão final da IA é: {decisao_final_ia}")
```
